```python

import json
import glob
import os
import pandas as pd
```

```python
files = os.listdir("../data/video/transcription")
transciption_length = {}
sorted_file = sorted(files)
# print(sorted_file)
for file in sorted_file:
    with open(os.path.join("../data/video/transcription",file)) as json_file:
        transcript = json.load(json_file)
        transciption_length[file] = len(transcript)
        print(file)
        print(len(transcript))
        
```

```python
imgfolders = os.listdir("../data/video/snapshot")#.remove('.DS_Store')
imgfolders.remove(".DS_Store")
snapshot_length = {}
sortted_imgfolder = sorted(imgfolders)
# print(sortted_imgfolder)
for imgfolder in sortted_imgfolder:
    snap_file_list = os.listdir(os.path.join("../data/video/snapshot",imgfolder))
    snapshot_length[imgfolder] = len(snap_file_list)
    print(imgfolder)
    print(len(snap_file_list))
        
```
